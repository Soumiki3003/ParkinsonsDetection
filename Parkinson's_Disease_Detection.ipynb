```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('parkinsons.csv')
df
```

### Lets visualize how many people has parkinsons (status - 1) and how many people doesn't (status - 0).

```python
import matplotlib.pyplot as plt
import seaborn as sns

temp = df['status'].value_counts()
temp_df = pd.DataFrame({'status' : temp.index , 'values': temp.values})
print(sns.barplot(x = 'status', y = 'values', data = temp_df))
```

There's a label imbalance, we need to fix this. 

```python
def distplots(i):
  sns.distplot(df[i])
  plt.show()

for i in list(df.columns)[1:]:
  distplots(i)
```

We are getting almost a nice Gaussian curve which means data is normally distributed.

```python
#Finding correlations

plt.figure(figsize = (15,15))
corr = df.corr()
sns.heatmap(corr, annot = True)
```

```python
X = df.drop(['status', 'name'], axis = 1)
y = df['status']
```

```python
#fixing the label imbalance

from imblearn.over_sampling import RandomOverSampler
from collections import Counter

before_fix = Counter(y)
ros = RandomOverSampler()
X_ros , y_ros = ros.fit_resample(X,y)
after_fix = Counter(y_ros)
print(f"Labels before fixing the imbalance : {before_fix}\nLabels after fixing the imbalance : {after_fix}")
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (-1,1))
x = scaler.fit_transform(X_ros)
y = y_ros
```

```python
# Applying feature engineering using PCA to deal with highly correlated features

from sklearn.decomposition import PCA

pca = PCA(n_components = 0.95)  #this means those min number of principal components will be chosen where 95% variance will be retained
X_pca = pca.fit_transform(x)

print(x.shape)
print(X_pca.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score, recall_score, precision_score, f1_score


from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(C = 0.4, max_iter=1000, solver = 'liblinear')
lr = classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_lr = accuracy_score(y_test, y_pred)

from sklearn.tree import DecisionTreeClassifier
classifier2 = DecisionTreeClassifier(random_state = 42)
dt = classifier2.fit(X_train, y_train)
y_pred2 = classifier2.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred2)

from sklearn.ensemble import RandomForestClassifier
classifier3 = RandomForestClassifier(random_state = 42)
rfe = classifier3.fit(X_train, y_train)
y_pred3 = classifier3.predict(X_test)
accuracy_rfe = accuracy_score(y_test, y_pred3)

from sklearn.svm import SVC
classifier4 = SVC(cache_size = 100)
svm = classifier4.fit(X_train, y_train)
y_pred4 = classifier4.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred4)

from sklearn.neighbors import KNeighborsClassifier
classifier5 = KNeighborsClassifier(n_neighbors = 3)
knn = classifier5.fit(X_train, y_train)
y_pred5 = classifier5.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred5)

from sklearn.naive_bayes import GaussianNB
classifier6 = GaussianNB()
gnb = classifier6.fit(X_train, y_train)
y_pred6 = classifier6.predict(X_test)
accuracy_gnb = accuracy_score(y_test, y_pred6)

from sklearn.naive_bayes import BernoulliNB
classifier7 = BernoulliNB()
gnb = classifier7.fit(X_train, y_train)
y_pred7 = classifier7.predict(X_test)
accuracy_bnb = accuracy_score(y_test, y_pred7)

list1 = ['Logistic Regression', 'Decision Tree Classifier', 'Random Forest Classifier', 'SVM', 'KNN', 'GaussianNB', 'BernoulliNB']
list2 = [accuracy_lr, accuracy_dt, accuracy_rfe,accuracy_svm, accuracy_knn, accuracy_gnb, accuracy_bnb ]
list3 = [classifier,classifier2,classifier3,classifier4,classifier5,classifier6,classifier7 ]

df_accuracy = pd.DataFrame({'Models ': list1, 'Accuracy': list2})
print(df_accuracy)

chart = sns.barplot(x = list1, y = list2, data = df_accuracy)
chart.set_xticklabels(chart.get_xticklabels(), rotation = 90)
print(chart)
```

```python
from xgboost import XGBClassifier

model_xgb = XGBClassifier()
model_xgb.fit(X_train, y_train)
y_pred_xgb = model_xgb.predict(X_test)
print(accuracy_score(y_test, y_pred_xgb)*100)
```

```python
cm = confusion_matrix(y_test, model_xgb.predict(X_test))
```

```python
f1_score(y_test, model_xgb.predict(X_test), average = 'binary')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, model_xgb.predict(X_test)))
print('Confusion Matrix :\n ', cm)
```
